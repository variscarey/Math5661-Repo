Consider Finite Difference approximations to $f^\prime(x)$, where 
$f(x) = \arctan{x}$.

Let $x=1$; hence $f(x)=\pi/4$ and $f^\prime(x)=1/2$.

The error from the forward difference is bounded by $\frac{h}{2}|f^{(2})(\xi)|$, $\xi\in[.5,.5+h] \approx $h/4$.

```python
import numpy as np
import matplotlib.pyplot as plt
from differences import *
%matplotlib inline

#j=np.arange(10,40,dtype='float')
#h=2**-j
sqreps=np.sqrt(np.finfo(float).eps)
h=np.linspace(1/8*sqreps,8*sqreps,100)
x=np.sqrt(2)
del2=(np.finfo(float).eps)**(1/3)
h2=np.linspace(1/32*del2,8*del2,100)

#err_est=
df1=for_diff(np.arctan,x,h)
df2=cent_diff(np.arctan,x,h2)
err1=np.absolute(1/3 - df1)*3
plt.plot(h,x*h/3,c='r')
err2=np.absolute(1/3 - df2)*3

plt.plot(h,err1)
plt.axvline(sqreps,c='r')
plt.figure()
plt.plot(h2,err2)
plt.axvline(del2,c='r')
plt.show()
```

Finite differencing noisy data is even more dangerous.
Let's add a little noise to $f(x)$.

```python
def per_f(x):
    return np.arctan(x)+1E-10*np.random.rand(1)


df1=for_diff(per_f,x,h)
df2=cent_diff(per_f,x,h2)
err1=np.absolute(1/3 - df1)*3
plt.plot(h,x*h/3,c='r')
err2=np.absolute(1/3 - df2)*3

plt.plot(h,err1)
plt.axvline(sqreps,c='r')
plt.figure()
plt.plot(h2,err2)
plt.axvline(del2,c='r')
plt.show()

h3=np.linspace(1E-6,1E-4,1000)
df1=for_diff(per_f,x,h3)
err3=np.absolute(1/3 - df1)*3
plt.plot(h3,err3)
```
